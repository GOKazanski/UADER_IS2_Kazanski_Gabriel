# Esfuerzo y Tiempo Calendario

```python

import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones para calcular el esfuerzo y el tiempo calendario
def calcular_esfuerzo(S):
    return 8 * (S ** 0.95)

def calcular_tiempo(E):
    return 2.4 * (E ** 0.33)

# Generar los valores de S (tamao) y calcular el esfuerzo E
S_vals = np.arange(0, 10001, 100)
E_vals = calcular_esfuerzo(S_vals)

# Generar los valores de E (esfuerzo) y calcular el tiempo calendario td
E_range = np.arange(1, 501, 1)
td_vals = calcular_tiempo(E_range)

# Crear las grficas
plt.figure(figsize=(12, 6))

# Grfico 1: Esfuerzo en funcin del tamao del proyecto
plt.subplot(1, 2, 1)
plt.plot(S_vals, E_vals, label='Esfuerzo (E)', color='b')
plt.xlabel('Tamao del proyecto (S)')
plt.ylabel('Esfuerzo (E)')
plt.title('Esfuerzo (E) en funcin del Tamao (S)')
plt.grid(True)

# Grfico 2: Tiempo calendario en funcin del esfuerzo
plt.subplot(1, 2, 2)
plt.plot(E_range, td_vals, label='Tiempo (td)', color='g')
plt.xlabel('Esfuerzo (E)')
plt.ylabel('Tiempo calendario (td)')
plt.title('Tiempo calendario (td) en funcin del Esfuerzo (E)')
plt.grid(True)

plt.tight_layout()
plt.show()
    
```
